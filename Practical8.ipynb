```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dt=sns.load_dataset('titanic')
```

```python
dt.head()
```

```python
sns.distplot(x = dt['age'], bins = 10)
```

```python
sns.distplot(x = dt['age'], bins = 10,kde = False)
```

```python
sns.jointplot(x = dt['age'], y = dt['fare'],kind = 'scatter')
```

```python
sns.jointplot(x = dt['age'], y = dt['fare'], kind = 'hex')
```

```python
sns.rugplot(dt['fare'])
```

```python
sns.barplot(x='sex',y='age', data=dt)
```

```python
sns.barplot(x='sex',y='age',data=dt,estimator=np.std)
```

```python
sns.countplot(x='sex',data=dt)
```

```python
sns.boxplot(x='sex', y='age', data=dt)
```

```python
sns.boxplot(x='sex', y='age', data=dt, hue="survived")
```

```python
sns.violinplot(x='sex', y='age', data=dt)
```

```python
sns.violinplot(x='sex', y='age', data=dt, hue='survived')
```

```python
sns.stripplot(x='sex', y='age', data=dt, jitter=False)
```

```python
sns.stripplot(x='sex', y='age', data=dt, jitter=True)
```

```python
sns.stripplot(x='sex', y='age', data=dt, jitter=True, hue='survived')
```

```python
sns.swarmplot(x='sex', y='age', data=dt)
```
