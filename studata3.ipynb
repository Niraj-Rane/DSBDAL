```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("studentdata.csv")
df
```

```python
df.isnull()
```

```python

#  Checking for missing values using isnull()
series = pd.isnull(df["math_score"]) 
df[series] 
```

```python
m_v=df['math_score'].mean()
df['math_score'].fillna(value=m_v,inplace=True)
df
```

```python
# Fill missing values as 0
ndf=df 
ndf.fillna(0)
```

```python
# Checking for missing values using notnull()
df.notnull()
```

```python
 # Filling a null values using replace() method 
ndf.replace(to_replace = np.nan, value = -99)
```

```python
# drop rows with at least 1 null value 
ndf.dropna() 
```

```python
# To Drop rows if all values in that row are missing 
ndf.dropna(how = 'all')
```

```python
m_v=df['writing_score'].mean()
df['writing_score'].fillna(value=m_v,inplace=True)
df
```

```python
m_v=df['placement_score'].mean()
df['placement_score'].fillna(value=m_v,inplace=True)
df
```

```python
col = ['math_score', 'reading_score' , 'writing_score','placement_score']
df.boxplot(col)
```
