```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dt=pd.read_csv("titanic.csv")
```

```python
dt
```

```python
dt.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis='columns',inplace=True)
```

```python
dt.head()
```

```python
dt.isnull().sum()
```

```python
m_v=dt['Age'].mean()
dt['Age'].fillna(value=m_v,inplace=True)
```

```python
dt
```

```python
dt.isnull().sum()
```

```python
target=dt.Survived
dt.drop(['Survived'],axis='columns',inplace=True)
```

```python
dt.head()
```

```python
dt['Sex'].replace({'female':1 ,'male':0},inplace=True)
```

```python
dt.corr()
```

```python
dt
```

```python
target
```

```python
X=dt[['Pclass','Sex','Age','Fare']]
Y=target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
X_train
```

```python
Y_train
```

```python
from sklearn.naive_bayes import GaussianNB 
gaussian = GaussianNB() 
gaussian.fit(X_train, Y_train) 
```

```python
Y_pred=gaussian.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
accuracy = accuracy_score(Y_test,Y_pred)
precision = precision_score(Y_test,Y_pred,average='micro')
recall = recall_score(Y_test,Y_pred,average='micro')
```
