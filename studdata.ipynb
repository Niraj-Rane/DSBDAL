```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("studentdata.csv")
df
```

```python
df.isnull()
```

```python
m_v=df['math_score'].mean()
df['math_score'].fillna(value=m_v,inplace=True)
df
```

```python
m_v=df['writing_score'].mean()
df['writing_score'].fillna(value=m_v,inplace=True)
df
```

```python
m_v=df['placement_score'].mean()
df['placement_score'].fillna(value=m_v,inplace=True)
df
```

```python
m_v=df['clubjoin_score'].mean()
df['clubjoin_score'].fillna(value=m_v,inplace=True)
df
```

```python
m_v=df['placementoffer_count'].mean()
df['placementoffer_count'].fillna(value=m_v,inplace=True)
df
```

```python
col = ['math_score', 'reading_score' , 'writing_score','placement_score']
df.boxplot(col)
```
